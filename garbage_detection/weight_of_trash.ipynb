```python
import google.generativeai as genai
import os

genai.configure(api_key="AIzaSyAdUAiOIinn5-lKYQxki4KNrkPkB38ejLs")
```

```python
model = genai.GenerativeModel("gemini-1.5-flash")
```

```python
myfile = genai.upload_file("samples/sample7.jpg")
```

```python
prompt='''Identify all objects that belong to the following classes in this image: Biodegradable, Plastic, Paper, Metal, Cardboard, Glass. Return the details of each detected object in a structured format, such as a list of dictionaries. Each dictionary should contain the following keys and values:
- "class": The class of the object [Biodegradable,Plastic,Paper,Metal,Cardboard,Glass]
- "weight": The estimated weight of the object in grams.

Example output:
[
    {"class": "Plastic", "weight": 200},
    {"class": "Paper", "weight": 150}
]
'''
result = model.generate_content([myfile, "\n\n",prompt])
print(result.text)
```

```python
import json
detected_objects = json.loads(result.text)
```

```python
for obj in detected_objects:
    object_class = obj["class"]
    object_weight = obj["weight"]
    print(f"Class: {object_class}, Weight: {object_weight} grams")
```
